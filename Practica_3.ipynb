```python
cali=int(input("Ingrese la calificación :"))
if cali >= 70:
  print("Felicidades aprobó")
  print(cali)
elif cali < 70 and cali > 50:
  print("Segunda oportunidad")
else:
  print("Reprobado :(")
```
